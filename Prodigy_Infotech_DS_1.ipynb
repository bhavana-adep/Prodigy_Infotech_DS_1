#**Trending Books**

```python
#importing packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#reading the dataset

df=pd.read_csv("/content/Top-100 Trending Books.csv")
df.head()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.shape
```

```python
book=df.groupby(['author','rating']).size().reset_index()
book=book.sort_values(by='rating',ascending=False)
book=book.head(7)
plt.figure(figsize=(9,6))
plt.xticks(fontsize=6)
plt.title("Top 7 books with respect to Ratings")
plt.xlabel("Author")
plt.ylabel("Ratings")
plt.bar(book['author'],book['rating'])
plt.show()
```
